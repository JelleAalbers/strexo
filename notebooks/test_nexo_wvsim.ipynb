```python
q = st.get_array(run_id, 'pad_waveforms', 
                 #config=dict(electron_lifetime=1 * strexo.units.us)
                )
```

```python
t_scale = strexo.units.us
plt.figure(figsize=(4, 10))
for i, w in enumerate(q[:200]):
    y = w['data']
    charge = np.sum(y)
    sum_abs = np.abs(y).sum()
    y = w['data'] + i
    t = (
        #w['time'] 
        + np.arange(len(y), dtype=int) * 10)
    plt.plot(t/t_scale, y, 
             c='r' if charge > 10 else ('b' if sum_abs > 1000 else 'g'),
             linewidth=1)

plt.xlabel("Time [us]")
plt.ylabel("Current [e/s] -- each pad is offset by its index")
plt.show()
```
